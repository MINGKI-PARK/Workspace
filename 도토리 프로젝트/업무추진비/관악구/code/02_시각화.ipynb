## 데이터 준비하기

```python
df_position = pd.read_csv('관악구 업무추진비 위치.csv')
df_position
```

## 지도시각화

```python
# 관악구청을 중심으로 객체 생성하기
m = folium.Map([37.478094, 126.951524], zoom_start=7)
m
```

```python
# 지도에 마커로 음식점 표시하기
for row in df_position.index:
    sub_lat =  df_position.loc[row, 'lat']
    sub_lng = df_position.loc[row, 'lng']
    
    title = df_position.loc[row,'사용장소']
    
    folium.Marker([sub_lat, sub_lng], tooltip=title).add_to(m)

m.save('sample.html')
m
```

### 히트맵으로 그려보기

```python
from folium.plugins import HeatMap
 
m = folium.Map(
    location = [36.5053542, 127.7043419],
    zoom_start = 8,
#     tiles = 'Cartodb Positron'
    tiles='stamentoner'
)

data = []

for row in df_position.index:
    
    sub_lat =  df_position.loc[row, 'lat']
    sub_lng = df_position.loc[row, 'lng']
    
    data.append([sub_lat, sub_lng])

# 관악구청 마커로 표시
folium.Marker([37.478094, 126.951524], tooltip='관악구청').add_to(m)
    

    
HeatMap(data).add_to(m)

m.save('heat.html')
m
```
