```python
for i in df.index:
    df.loc[i, '지출'] = df.loc[i, '지출'].replace(',', '')
    df.loc[i, '노출'] = df.loc[i, '노출'].replace(',', '')
    df.loc[i, '클릭'] = df.loc[i, '클릭'].replace(',', '')
    df.loc[i, '웹 사용자'] = df.loc[i, '웹 사용자'].replace(',', '')
    df.loc[i, 'CVR'] = df.loc[i, 'CVR'][:-1]
    df.loc[i, '매출'] = df.loc[i, '매출'].replace(',', '')
```

```python
df['지출'] = df['지출'].astype('int')
df['노출'] = df['노출'].astype('int')
df['클릭'] = df['클릭'].astype('int')
df['웹 사용자'] = df['웹 사용자'].astype('int')
df['CVR'] = df['CVR'].astype('float')
df['매출'] = df['매출'].astype('int')
```
