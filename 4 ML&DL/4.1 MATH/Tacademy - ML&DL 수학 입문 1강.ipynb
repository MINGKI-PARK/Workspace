```python
np.random.seed(0)

A1 = np.random.random(100)
A2 = np.random.random(100)
A1.sort()
A2.sort()
print(A1.shape)
```

```python
m = -0.5 # 평균
s = 0.3 # 편차

A1 = np.random.normal(m, s, 250)
A1.sort()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

ax.scatter(A1, A1, c='b', marker='.')
plt.show()
```
