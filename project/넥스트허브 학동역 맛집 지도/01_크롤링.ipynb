## 크롤링

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.implicitly_wait(3)

# 카카오 지도로 이동
driver.get('https://map.kakao.com/')
time_for_web()

# '논현동 음식점' 검색어 입력
driver.find_element_by_id('search.keyword.query').send_keys('논현동 음식점')
driver.implicitly_wait(3)

# 검색 클릭
try:
    searchbutton = driver.find_element_by_xpath("//button[@id='search.keyword.submit']") # 검색버튼 누르기
    driver.execute_script("arguments[0].click();", searchbutton)
    time.sleep(2)
except:
    print('확인 필요')

    
# 더보기 클릭
try:
    morebutton = driver.find_element_by_xpath('//*[@id="info.search.place.more"]') # 더보기 버튼
    driver.execute_script("arguments[0].click();", morebutton)
    time.sleep(2)
except:
    print('확인 필요')
    
# 1페이지 버튼 클릭
pagebutton = driver.find_element_by_xpath('//*[@id="info.search.page.no1"]') # 1page
driver.execute_script("arguments[0].click();", pagebutton)
```

```python
restaurant = {
    'name': [], 
    'category': [], 
    'address': [],
    'score': [], 
    'eval_cnt': [], 
    'review_cnt': []
}
```

```python
for pagenum in tqdm(range(1, 35)):
    try: 
        # 현재 페이지의 음식점 정보 크롤링
        page = pagenum % 5
        soup = driver.page_source
        b = BeautifulSoup(soup, 'lxml')
        for num in range(0, 15):
            restaurant['name'].append(b.find_all('a','link_name')[num].text)
            restaurant['category'].append(b.find_all('span', 'subcategory')[num].text)
            restaurant['address'].append(b.find_all('p', 'lot_number')[num].text)
            restaurant['score'].append(b.find_all('em', 'num')[num].text)
            restaurant['eval_cnt'].append(b.find_all('a', 'numberofscore')[num].text)
            restaurant['review_cnt'].append(b.find_all('a', 'review', 'em')[num].text)
        time_for_web
        
        # 다음 페이지 or 다음 버튼
        if page:
            pagebutton = driver.find_element_by_id('info.search.page.no{}'.format(page+1))
#             pagebutton = driver.find_element_by_xpath('//*[@id="info.search.page.no1"]') # page
            driver.execute_script("arguments[0].click();", pagebutton)
        else:
            nextbutton = driver.find_element_by_id('info.search.page.next')
            driver.execute_script("arguments[0].click();", nextbutton)
    except:
        print('페이지 초과 or 끝')
        break
```
