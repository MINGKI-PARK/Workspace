## 데이터 합치기

### 의장단 월별 데이터 합치기

```python
# 의장단 데이터 합치기
df1 = pd.concat([d1, d2, d3, d4, d5, d6, d7, d8, d9, d10, d11, d12], ignore_index=True)
df1.head()
```

### 의회사무국 월별 데이터 합치기

```python
# 의회사무국 데이터 합치기
df2 = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)
df2.head()
```

### 의장단, 사무국 데이터 합치기

```python
df1.shape, df2.shape
```

```python
df1.columns, df2.columns
```

* 컬럼 이름 맞춰주고 필요없는 컬럼은 삭제하자

#### 컬럼 이름 변경하기

```python
# 의장단 컬럼 이름 변경
df1.rename(columns={'Unnamed: 3' : '사용시간'}, inplace=True)
```

* Unnamed: 3 --> 사용시간으로 변경

```python
# 사무국 컬럼 이름 변경
df2.rename(columns={'Unnamed: 3' : '사용시간',
                    '집행방법' : '사용방법'}, inplace=True)
```

#### 컬럼 추가하기

```python
# 사용자 컬럼 추가하기
df2['사용자'] = '의회사무국'
df2.head()
```

```python
df1.shape, df2.shape
```

#### 컬럼 삭제하기

```python
# 사무국에서 예산과목 컬럼 삭제하기
del df2['예산과목']
df2.head()
```

#### 합치기

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```
